```python
# Import necessary libraries
import pandas as pd
from sklearn.datasets import load_iris

# Load the Iris dataset
iris = load_iris()
df = pd.DataFrame(iris.data, columns=iris.feature_names)
df['species'] = iris.target

# Display the first few rows of the dataset
df.head()
```

```python
# Check data types and missing values
df.info()

# Check for any missing values
df.isnull().sum()
```

```python
# Handle missing data (if there were any)
df.fillna(df.mean(), inplace=True)  # Fill missing values with the mean of each column
```

```python
# Basic statistics of numerical columns
df.describe()
```

```python
# Group by species and calculate the mean of each numerical feature
species_group = df.groupby('species').mean()
species_group
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Bar chart: average petal length by species
sns.barplot(x='species', y='petal length (cm)', data=df)
plt.title('Average Petal Length by Species')
plt.xlabel('Species')
plt.ylabel('Petal Length (cm)')
plt.show()
```

```python
# Histogram of petal length
plt.hist(df['petal length (cm)'], bins=10, color='skyblue', edgecolor='black')
plt.title('Distribution of Petal Length')
plt.xlabel('Petal Length (cm)')
plt.ylabel('Frequency')
plt.show()
```

```python
# Scatter plot: sepal length vs petal length
plt.scatter(df['sepal length (cm)'], df['petal length (cm)'], color='green')
plt.title('Sepal Length vs Petal Length')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Petal Length (cm)')
plt.show()
```
